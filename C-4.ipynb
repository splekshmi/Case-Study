### 1) Mean sales generated by each zone

```python
Ms=Sales.mean(numeric_only=True).round(3)
Mean_Sales=pd.DataFrame(Ms,columns=['Mean sales'])          # to give desired column names,creating a new data frame
Mean_Sales
```

### 2) Total sales generated by all the zones for each month

```python
ts=Sales.sum(numeric_only=True,axis=1)
Total_Sales=pd.DataFrame({'Month':Sales['Month'],'Total Sales':ts})  # to give desired column names,creating a new data frame
Total_Sales.style.hide_index()
```

### 3)Check whether all the zones generate the same amount of sales

```python
# Stating the hypothesis
```

H0= All the zones generate the same amount of sales,µ1=µ2=µ3

Ha= Atleast one zone generate sales amount that differs from the rest,All μ are not equal

```python
# setting the significance level
```

Considering significance level(α)=0.05 or 5%

```python
# Determining p value using one way anova (for comparing means from 4 zones)
```

```python
import numpy as np
from scipy.stats import f_oneway
a=np.array(Sales['Zone - A'])
b=np.array(Sales['Zone - B'])
c=np.array(Sales['Zone - C'])
d=np.array(Sales['Zone - D'])
f_oneway(a,b,c,d)
```

p value: 0.0011827601694503335≈ 0.001< 0.05

Rejecting the null hypothesis

```python
#Conclusion
```

Since p value is less than the selected significance level of 0.05 ,rejecting the null hypothesis which means that atleast one zone generate sales amount that differs from the rest(All μ are not equal)
